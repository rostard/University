#Критерій Колмогорова
Будемо моделювати рівномірний розподіл на проміжку [low, high]

```
n = 100

low = 140
high = 150

diff = high - low

#функція розподілу для рівномірного розподілу
def eq_dist(x, low, high):
  if x < low:
    return 0
  if x > high:
    return 1
  return (x - low) / (high - low)
```

```
data1 = [(low + high) / 2 + np.random.randint(-diff//2, diff//2) for _ in range(n)] #генеруємо данні
```

```
guess_dist = [eq_dist(low + x*(high - low) /  n, low, high) for x in range(n)]
```

```
sample_dist = [eq_dist(x, low, high) for x in sorted(data1)]
```

```
D = max(abs(guess_dist[i] - sample_dist[i]) for i in range(n))
z = D * np.sqrt(n)
```

For** a=0.05** we find from table **z<sub>1–α</sub>=1,36**

```
if z < 1.36:
  print("Гіпотезу не відхиляємо")
else:
  print("Гіпотезу відхиляємо")
```

#Критерій Вілкоксона

```
n = 90
m = 100
# Генеруємо данні
data1 = [150 + np.random.randint(-10, 10) for _ in range(n)]
data2 = [160 + np.random.randint(-10, 10) for _ in range(m)]
```

```
last = 0
ranks = dict()

#розраховуємо об'єднанні ранги
for x in sorted((set(data2) | set(data1))):
  c = (data1+data2).count(x)
  print(x, c, end=' ')
  print(last + (c+1)/2)
  last += c
  ranks[x] = last + (c+1)/2
  
```

```
W=0
for x in data1:
  W+=ranks[x]
print(W)
```

```
#Розраховуэмо критичні значення W
low_W = 0.5*n*(n+m+1)  - 2 *  np.sqrt(1/12 * n * m * (n + m + 1))
high_W = n * (n+m+1) - low_W
```

```
print(low_W, high_W)
```

```
if low_W < W < high_W:
  print("Гіпотезу не відхиляємо")
else:
  print("Гіпотезу відхиляємо")
```
