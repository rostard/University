Функція для генерації ранжування розмірності n

```
n = 10
def genPriorities(n):
  l = 1
  ret = list()
  for i in range(n):
    ret.append(l)
    if np.random.rand() > 0.5:
      l += 1
  np.random.shuffle(ret)
  return ret
    
```

```
#Функція для встановлення об'єднаних рангів
def prepareRanks(A):
  l = 1
  
  d = dict()
  
  for i in set(A):
    n = A.count(i)
    
    d[i] = (l + l+n-1) / 2
    
    l+=n
  ret = list()
  for i in range(len(A)):
    ret.append(d[A[i]])
  

def T(A):
  T_A = 0
  for i in set(A):
    n = A.count(i)
    T_A += n**3 - n
  T_A *= (1/12)
  return T_A

def Spirmen(A, B):
  n = len(A)
  
  T_A = T(A)
  T_B = T(B)
  
  print("T_A", T_A)
  print("T_B", T_B)
  
  prepareRanks(A)
  prepareRanks(B)
  
  s = 0
  for i in range(n):
    s += (A[i] - B[i]) ** 2
  
  print("s", s)
  
  nom = (1/6) * (n**3 - n) - T_A - T_B - s
  denom = np.sqrt(((1/6)*(n**3-n) - 2*T_A) * ((1/6)*(n**3-n) - 2*T_B))
  
  return nom/denom
```
