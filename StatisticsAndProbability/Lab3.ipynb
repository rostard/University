Вихідні дані змоделювати у вигляді 5-ти груп, кожна з яких відповідає за певний ступінь дії фактора(дані в кожній групі моделюються як певне фіксоване значення плюс або мінус невелике по відношенню до початкового значення випадкове число). 

```
import scipy.stats
import numpy as np
```

```
p = 5 #кількість груп
ns = np.random.randint(8, 12, (p, 1)) #Кількість елементів кожної групи
N = sum(ns)

#degrees of freedom
k1 = N-p
k2 = p-1
```

```
data = []
for n in ns:
  center = np.random.randint(5, 50) #певне центральне значення [5, 49]
  data.append(center + np.random.randint(-5, 6, (int(n), 1))) #плюс відхилення [-5, 5]

# print("Data:")
# for row in data:
#   print(row)
# print()

#рахуємо середні значення
means = []
total_mean = 0
for group in data:
  total_mean += np.sum(group)
  means.append(np.mean(group))
total_mean = total_mean / N

print(means)
print(total_mean)
```

```
#рахуємо дисперсії
s1 = 0
for i in range(len(n)):
  s1 += sum(np.square(data[i]-means[i]))
s1 /= k1
print("s1:", s1)


s2 = 0
for i in range(len(n)):
  s2 += ns[i]*np.square(means[i]-total_mean)
s2 /= k2
print("s2:", s2)
```

Розраховуємо випадкову величину F, що має розподіл Фішера–Снедекора

Далі розраховуємо ймовірність отримати значення не менш екстримальне ніж це. 

```
F = (s2 / s1) * (k2 / k1)
print("F:", F)
p = 1 - scipy.stats.f.cdf(F, k1, k2)
print("p:", p)
```

```
if p < 0.05:
  print("Вплив є істотним")
else:
  print("Вплив не є істотним")
```
